## Running each model against each core dataset 


```python
import os
import sys

root_path = os.path.abspath(os.path.join(os.getcwd(), "../.."))
sys.path.append(root_path)

import time

import torch
import pandas as pd
import matplotlib.pyplot as plt
import pytorch_lightning as pl
from pytorch_lightning.loggers import CSVLogger
from pytorch_lightning.callbacks import (
    LearningRateMonitor,
    ModelCheckpoint,
    EarlyStopping,
)
from src.utils.helpers import load_config
from src.training.dataset import ImageDataModule
from src.models.classification_model import ImageClassifier
```

```python
trainer = pl.Trainer(accelerator="auto", devices="auto")

ckpts = [
    "../Clip-Vit Standard/checkpoints/last.ckpt",
    "../Clip-Vit Cropped/checkpoints/last.ckpt",
    "../Clip-Vit Scaled-Lan/checkpoints/last.ckpt",
    "../Clip-Vit Scaled-BiCubic/checkpoints/last.ckpt",
    "../Clip-Vit Scaled-BiLinear/checkpoints/last.ckpt",
    "../ConvNext Standard/checkpoints/last.ckpt",
    "../ConvNext Cropped/checkpoints/last.ckpt",
    "../ConvNext Scaled/checkpoints/last.ckpt",
    "../ConvNext Scaled-BiCubic/checkpoints/last.ckpt",
    "../ConvNext Scaled-BiLinear/checkpoints/last.ckpt",
    "../EfficientNet Standard/checkpoints/last.ckpt",
    "../EfficientNet Cropped/checkpoints/last.ckpt",
    "../EfficientNet Scaled/checkpoints/last.ckpt",
    "../EfficientNet Scaled-BiCubic/checkpoints/last.ckpt",
    "../EfficientNet Scaled-BiLinear/checkpoints/last.ckpt"
]
```

```python
# Standard Dataset
config = load_config("sampled_clip_vit.yaml") # This can really be anything
config1 = config.copy()
config1["data"]["test_path"] = "../../src/data/sampled/test"
data_module = ImageDataModule(config1)

for ckpt_path in ckpts:
    model = ImageClassifier.load_from_checkpoint(ckpt_path)
    model.eval()    
    model.freeze() 

    results = trainer.test(model=model, datamodule=data_module)
    print(ckpt_path, results)
```
