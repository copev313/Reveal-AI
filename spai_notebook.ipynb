```python
import warnings
import pandas as pd

warnings.filterwarnings("ignore")
```

## Unaltered SPAI Results

```python
sampled_ai_results_df = pd.read_csv("src/logs/spai_outputs/sampled-ai.csv")
sampled_ai_results_df["true_label"] = 1
sampled_real_results_df = pd.read_csv("src/logs/spai_outputs/sampled-real.csv")
sampled_real_results_df["true_label"] = 0

# Combine rows from both dataframes:
combined_df = pd.concat([sampled_ai_results_df, sampled_real_results_df], ignore_index=True)

# Add predicted label:
combined_df["pred_label"] = 0
combined_df["pred_label"].loc[combined_df["spai"] > 0.5] = 1
combined_df
```

```python
# Histogram of real data:
sampled_real_results_df["spai"].plot.hist(bins=20, alpha=0.5, color="blue")
```

```python
# Histogram of AI-generated data:
sampled_ai_results_df["spai"].plot.hist(bins=20, alpha=0.5, color="orange")
```

```python
# Calc accuracy:
accuracy = (combined_df["true_label"] == combined_df["pred_label"]).mean()
print(f"SPAI Accuracy on TEST SET: {accuracy:.4f}")
```
